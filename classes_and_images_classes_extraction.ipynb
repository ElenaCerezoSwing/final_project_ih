```python
import os
import re
import pandas as pd
```

```python
classes = os.listdir('./data/input_images')
images_paper = os.listdir('./data/input_images/paper')
images_cardboard = os.listdir('./data/input_images/cardboard')
classes
```

```python
def get_picture_per_classes(classes):
    return [os.listdir('./data/input_images/'+ item) for item in classes]
```

```python
images = get_picture_per_classes(classes)
images
```

```python
def get_image_name(item):
    return item.split('.')[0]
```

```python
images_name = [get_image_name(item) for image in images for item in image]
```

```python
images_name
```

```python
def get_classes_from_images(item):
    num = re.findall('\d+', item)
    return item.split(num[0])[0]
```

```python
s = get_classes_from_images('elena2')
s
```

```python
image_classes = [get_classes_from_images(item) for item in images_name]
```

```python
image_classes
```

```python
new_dict = {'Images_name':images_name, 'Image_class': image_classes}
```

```python
p = pd.DataFrame(new_dict, index=images_name, columns=['Image_class'])
```

```python
p.head()
```

```python
new_p = pd.get_dummies(p)
```

```python
new_p.columns
```

```python
new_p = new_p.rename(columns={'Image_class_cardboard':'cardboard', 'Image_class_glass':'glass', 'Image_class_metal':'metal',
       'Image_class_paper':'paper', 'Image_class_plastic':'plastic', 'Image_class_trash':'trash'})
new_p.head()
```
